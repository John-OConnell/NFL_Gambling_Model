```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
from vardata import years, team_abbr, columns, teams_idx
from results_processor import *
```

```python
# read in data csv to df
df = pd.read_csv('stats/final_data.csv', index_col=[0])
```

```python
# preview the data to ensure it was read correctly
print('Dimension of dataset= ', df.shape)
df.head() # To get first n rows from the dataset default value of n is 5
```

```python
X = df.values[:, 2:70]  # get input values from first two columns
y = df.values[:, 73]  # get output values from last coulmn
m = len(y) # number of training examples
```

```python
# preview the data
print('Total no of samples (m) = %s \n' %(m))

# Show only first 5 records
for i in range(5):
    print('X =', X[i, ], ', y =', y[i])
```

```python
# split data into train and test sets
# 10 years of data for training, 3 years of data for testing
X_train = X[:2560]
X_test = X[2560:]
y_train = y[:2560]
y_test = y[2560:]
```

*************************************************************
**Testing to find best epsilon to use for Huber Loss:**
*************************************************************

```python
win_p = []
rmse = []
epsilons = []
# looping through a range of epsilon values
# all work done here comes from results_processor.py
for i in range(100, 401):
    epsilon = i/100
    model_r = linear_model.HuberRegressor(epsilon = epsilon, max_iter = 10000)
    scaler = preprocessing.MinMaxScaler().fit(X_train)
    X_scaled_train = scaler.transform(X_train)
    X_scaled_test = scaler.transform(X_test)
    model_r.fit(X_scaled_train,y_train)
    
    results = np.array([])
    
    for each in X_scaled_test:
        outcome = model_r.predict([each])
        results = np.append(results, [outcome])
        
    results_df = pd.DataFrame(columns = ['ht_pts_diff', 'ht_vegas_line', 'model_pred'])
    results_df['ht_pts_diff'] = df['ht_pts_diff'][2560:]
    results_df['ht_vegas_line'] = df['ht_vegas_line'][2560:]
    results_df = results_df.reset_index(drop = True)
    results_df['model_pred'] = results.tolist()
    
    count = 0
    no_bet = 0
    for index in results_df.index:
        ht_pts_diff = results_df['ht_pts_diff'][index]
        vegas_line = results_df['ht_vegas_line'][index]
        model_pred = round(results_df['model_pred'][index])
        if model_pred > 0 and vegas_line <= 0:
            if model_pred > abs(vegas_line) and ht_pts_diff > abs(vegas_line):
                count+=1
            elif model_pred < abs(vegas_line) and ht_pts_diff < abs(vegas_line):
                count+=1
            elif model_pred == abs(vegas_line):
                no_bet+=1
        elif model_pred > 0 and vegas_line > 0:
            if ht_pts_diff > 0 or abs(ht_pts_diff) < vegas_line:
                count+=1
        elif model_pred < 0 and vegas_line >= 0:
            if abs(model_pred) < vegas_line and (ht_pts_diff > 0 or abs(ht_pts_diff) < vegas_line):
                count+=1
            elif abs(model_pred) > vegas_line and ht_pts_diff < 0 and abs(ht_pts_diff) > vegas_line:
                count+=1
            elif abs(model_pred) == vegas_line:
                no_bet+=1
        elif model_pred < 0 and vegas_line < 0:
            if ht_pts_diff < 0 or ht_pts_diff < abs(vegas_line):
                count+=1
        elif model_pred == 0:
            if vegas_line > 0 and (ht_pts_diff > 0 or abs(ht_pts_diff) < vegas_line):
                count+=1
            if vegas_line < 0 and (ht_pts_diff < 0 or ht_pts_diff < abs(vegas_line)):
                count+=1
        else:
            print('something not accounted for')
            print(ht_pts_diff, vegas_line, model_pred)

    temp_win_p = count/(len(results_df.index) - no_bet)
    temp_rmse = mean_squared_error(results_df['ht_pts_diff'], results_df['model_pred'], squared = False)
    win_p.append(temp_win_p)
    rmse.append(temp_rmse)
    epsilons.append(epsilon)
```

```python
# function to annotate the maximum value on a graph
def annot_max(x,y, ax=None):
    xmax = x[np.argmax(y)]
    ymax = max(y)
    text= "x={:.3f}, y={:.3f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->")#,connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.96), **kw)

# function to annotate the minimum value on a graph
def annot_min(x,y, ax=None):
    xmin = x[np.argmin(y)]
    ymin = min(y)
    text= "x={:.3f}, y={:.3f}".format(xmin, ymin)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmin, ymin), xytext=(0.94,0.06), **kw)
```

```python
# plotting Winning Percentage vs Huber Loss Epsilon
plt.figure(figsize=(8,6))
plt.plot(epsilons, win_p)
plt.title("Winning Percentage vs Huber Loss Delta")
plt.xlabel("Delta")
plt.ylabel("Winning Percentage")

annot_max(epsilons, win_p)
```

```python
# plotting Root Mean Squared Error vs Huber Loss Epsilon
plt.figure(figsize=(8,6))
plt.plot(epsilons, rmse)
plt.title("Root Mean Squared Error vs Huber Loss Delta")
plt.xlabel("Delta")
plt.ylabel("Root Mean Squared Error")

annot_min(epsilons, rmse)
```
